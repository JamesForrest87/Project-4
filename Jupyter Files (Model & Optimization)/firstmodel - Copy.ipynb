```python
import pandas as pd
from sqlalchemy import create_engine, text

# Create engine to male_fertility.sqlite
file_path = "/content/drive/MyDrive/Colab Notebooks/male_fertility.sqlite"
engine = create_engine(f"sqlite:///{file_path}")

# Define your SQL query
query = text("SELECT * FROM male_sperm")

# Execute the query and load data into a DataFrame
with engine.connect() as connection:
    result = connection.execute(query)
    # Fetch all rows into a list of tuples
    rows = result.fetchall()

# Get the column names from the result set
columns = result.keys()

# Create a DataFrame from the rows and columns
male_sperm = pd.DataFrame(rows, columns=columns)

# Print the DataFrame
print(male_sperm)




```

```python
male_sperm = male_sperm.drop(columns = ['Religion'], axis =1)
male_sperm
```

```python
# Choose a cutoff value and create a list of Age types to be replaced
# use the variable name `Age_types_to_replace`
Age_types_to_replace = list(counts[counts<10].index)

# Replace in dataframe
for app in Age_types_to_replace:
    male_sperm['Spermquality'] = male_sperm['Spermquality'].replace(app,"Other")

# Check to make sure binning was successful
male_sperm['Spermquality'].value_counts()
```

```python
male_sperm = pd.get_dummies(male_sperm, dtype=float)
male_sperm.head()
```

```python
y = male_sperm['No.ofchildren'].values
y
```

```python
X = male_sperm.drop('No.ofchildren', axis =1).values
X
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
input_features = len(X_train_scaled[0])
first_hidden_layer = 20
second_hidden_layer = 20
third_hidden_layer = 8

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=first_hidden_layer, input_dim = input_features, activation='relu'))
# Second hidden layer
nn.add(tf.keras.layers.Dense(units=second_hidden_layer, activation = 'relu'))
# Third hidden layer
nn.add(tf.keras.layers.Dense(units=third_hidden_layer, activation = 'relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1,activation = 'sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train,validation_split = 0.15, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
